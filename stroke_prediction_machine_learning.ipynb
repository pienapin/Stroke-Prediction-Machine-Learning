# Import Library and Dataset

```python
import pandas as pd
```

```python
df = pd.read_csv("/content/MImamAlvienCennaAlnaza.csv")
df.head()
```

# Data Preparation

```python
df.info()
```

```python
df.isnull().sum( )
```

```python
df = df.dropna(subset=['bmi'])
```

```python
df.isnull().sum( )
```

```python
df.drop('id', inplace=True, axis=1)
```

```python
df.info()
```

```python
categorical = ['gender', 'hypertension', 'heart_disease', 'ever_married', 'work_type',  'Residence_type', 'smoking_status']
```

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
```

```python
df[categorical] = df[categorical].apply(le.fit_transform)
```

```python
df.info()
```

```python
df.sample(5)
```

# Model Comparison

```python
!pip install pycaret shap # untuk menginstalasi pycaret dan shap
```

```python
from pycaret.classification import * 
```

```python
testing = setup(df, target='stroke', categorical_features=categorical)
```

```python
best = compare_models()
```

```python
test = create_model("dt")
```

```python
type(best)
```

```python
type(test)
```

```python
df.head()
```

```python
tuned_best = tune_model(best)
```

```python
tuned_test = tune_model(test)
```

Karena setelah dituning model memberikan F1 Score yang rendah, maka model yang akan dicoba untuk digunakan adalah model yang tidak dituning

```python
predict_model(best, df[df['stroke'] == 1].sample(5))
```

```python
predict_model(test, df[df['stroke'] == 1].sample(5))
```

```python
predict_model(test, df[df['stroke'] == 0].sample(5))
```
